<b>finding and modfying or removing outliers<b>

```python
I_q1=df['Inches'].quantile(0.25)
I_q3=df['Inches'].quantile(0.75)
INCH_IQR=I_q3-I_q1
```

```python
INCH_outliers=df[(df['Inches'] < (I_q1 - 1.5 * INCH_IQR)) | (df['Inches'] > (I_q3 + 1.5 * INCH_IQR))]
```

```python
Inch_out_index=INCH_outliers[INCH_outliers['Inches']>19].index
```

```python
df.loc[Inch_out_index,'Inches']=df.loc[Inch_out_index,'Inches'].apply(lambda x : 19)
```

```python
W_Q1=df['Weight'].quantile(0.25)
w_Q3=df['Weight'].quantile(0.75)
Weight_IQR=W_Q3-W_Q1
```

```python
Weight_outliers=df[(df['Weight']<(W_Q1-1.5*Weight_IQR)) | (df['Weight']>(W_Q3 +1.5*Weight_IQR))]
```

```python
Weight_out_index=df.loc[(df['Weight'] <= 0.7) | (df['Weight'] >= 5)].index

```

```python
df.loc[Weight_out_index,'Weight']=df.loc[Weight_out_index,'Weight'].apply(lambda x : 2)
```

<b>finding corelations between the numeric columns and category columns<b>

```python
num_columns=['Inches','Ram','Price','Weight']
cat_columns=['Company', 'TypeName', 'Cpu', 'Gpu', 'OpSys']
```

```python
corr_col=df[num_columns].corr()
```

```python
corr_col
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_col, cmap='RdYlGn', annot=True,center=0.4)
plt.show()
```

```python
for category in  cat_columns:
    for number in num_columns:
        plt.figure(figsize=[20,5])
        plt.title(f'bar graph of {category} v/s {number}')
        sns.barplot(data=df,x=category,y=number, palette="viridis")
        plt.grid(axis='y')
        plt.xticks(rotation=90)
```
